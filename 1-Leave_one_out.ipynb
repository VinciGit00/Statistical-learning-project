```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
pd.set_option("display.max_rows", 999)
df = pd.read_csv("Dataset/universal_top_spotify_songs.csv")
df.head()
```

```python
# Seleziona solo le colonne numeriche (features)
X = df.drop(['spotify_id', 'name', 'artists', 'country', 'snapshot_date', 'daily_rank'], axis=1)

# Target
y = df['daily_rank']

# Calcola il numero totale di iterazioni
total_iterations = len(X)

# Inizializza una lista per registrare i punteggi
scores = []

# Itera su ogni riga del DataFrame
for i in range(total_iterations):
    # Seleziona una riga come dato di test e il resto come dati di addestramento
    X_train = X.drop(i)
    y_train = y.drop(i)
    X_test = X.iloc[[i]]
    y_test = y.iloc[[i]]
    
    # Costruisci il modello polinomiale con regressione lineare
    model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
    
    # Valuta il modello utilizzando la cross-validation
    score = cross_val_score(model, X_train, y_train, cv=len(X_train), scoring='neg_mean_squared_error')
    
    # Aggiungi il punteggio alla lista
    scores.append(score.mean())
    
    # Calcola la percentuale di avanzamento
    progress_percentage = ((i + 1) / total_iterations) * 100
    
    # Stampa l'avanzamento
    print(f"Progress: {progress_percentage:.2f}%")

# Calcola la media dei punteggi
mean_score = sum(scores) / len(scores)
print("Mean Negative Mean Squared Error:", mean_score)
```
