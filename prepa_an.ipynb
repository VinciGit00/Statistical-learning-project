```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler 
```

```python
df = pd.read_csv("houses_madrid.csv")
df.head()
```

```python
df.dtypes
```

```python
# Replace non-numeric values with 0
df['bath_num'] = df['bath_num'].replace(['bath_num', 'sin baños'], 0)

# Convert the column to integer type
df['bath_num'] = df['bath_num'].astype(int)
```

```python
df["bath_num"] = df["bath_num"].fillna(0)

# Convert the column to integer type
df["bath_num"] = df["bath_num"].astype(int)
```

```python
non_numeric_values = df[~df['bath_num'].astype(str).str.isdigit()]['bath_num']
print(non_numeric_values)
```

```python
non_numeric_values = df[~df['built_in_wardrobe'].astype(str).str.isdigit()]['built_in_wardrobe']
print(non_numeric_values)
```

```python
# Replace non-numeric values with 0
df['built_in_wardrobe'] = df['built_in_wardrobe'].replace(['built_in_wardrobe'], 0)

# Convert the column to integer type
df['built_in_wardrobe'] = df['built_in_wardrobe'].astype(int)
```

```python
non_numeric_values = df[~df['chimney'].astype(str).str.isdigit()]['chimney']
print(non_numeric_values)
```

```python
df['chimney'] = df['chimney'].replace(['chimney'], 0)
df['chimney'] = df['chimney'].astype(int)

df['garden'] = df['garden'].replace(['garden'], 0)
df['garden'] = df['garden'].astype(int)

df['balcony'] = df['balcony'].replace(['balcony'], 0)
df['balcony'] = df['balcony'].astype(int)

df['construct_date'] = df['construct_date'].replace(['construct_date'], 0)
df['construct_date'] = df['construct_date'].astype(float) 

df['kitchen'] = df['kitchen'].replace(['kitchen'], 0)
df['kitchen'] = df['kitchen'].astype(float) 

df['lift'] = df['lift'].replace(['lift'], 0)
df['lift'] = df['lift'].astype(float) 


df['m2_real'] = df['m2_real'].replace(['m2_real'], 0)
df['m2_real'] = df['m2_real'].astype(float) 

df['m2_useful'] = df['m2_useful'].replace(['m2_useful'], 0) 
df['m2_useful'] = df['m2_useful'].astype(float) 


df['reduced_mobility'] = df['reduced_mobility'].replace(['reduced_mobility'], 0)
df['reduced_mobility'] = df['reduced_mobility'].astype(float) 

df['room_num'] = df['room_num'].replace(['room_num', 'sin habitación'], 0)
df['room_num'] = df['room_num'].astype(float) 

df['storage_room'] = df['storage_room'].replace(['storage_room'], 0)
df['storage_room'] = df['storage_room'].astype(float) 

df['swimming_pool'] = df['swimming_pool'].replace(['swimming_pool'], 0)
df['swimming_pool'] = df['swimming_pool'].astype(float) 

df['terrace'] = df['terrace'].replace(['terrace'], 0)
df['terrace'] = df['terrace'].astype(float) 

df['price'] = df['price'].replace(['price'], 0)
df['price'] = df['price'].astype(float) 
```

```python
df
```

```python
df = df.drop(["ad_description", "ad_last_update", "condition", "energetic_certif", "floor", 
            "garage", "loc_district", "loc_full", "loc_neigh", "loc_street", "loc_zone", "orientation",
            "house_type", "obtention_date", "air_conditioner", "house_id", "loc_city", "ground_size", 
            "heating", "unfurnished", "kitchen"], axis=1)
```

```python
df = df.dropna()
```

```python
y = df["price"]
df = df.drop("price", axis=1)
```

```python
df = df.dropna()
```

```python
X = df
print(X.shape)
X
```

```python
reg = LinearRegression().fit(X, y)
reg.score(X, y)
```
